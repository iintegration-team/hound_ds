````python
import re

def clean_text(text):
    # Remove excessive dots, normalize line breaks, and collapse multiple spaces
    text = re.sub(r'\.+', '', text)  # Remove long sequences of dots
    text = re.sub(r'[\r]+', '', text)  # Normalize line breaks
    text = re.sub(r'\s+', ' ', text)  # Collapse multiple spaces into one
    text = text.strip()  # Trim whitespace at start and end

    # Remove unwanted phrases
    unwanted_phrases = [
        '–°–∫—Ä—ã—Ç—å', '–ü–æ–ª–∏—Ç–∏–∫–∞ –∫–æ–Ω—Ñ–∏–¥–µ–Ω—Ü–∏–∞–ª—å–Ω–æ—Å—Ç–∏', '–û–±—Ä–∞—Ç–Ω–∞—è —Å–≤—è–∑—å', '¬©', '–ü–æ–¥—Ä–æ–±–Ω–µ–µ'
    ]
    for phrase in unwanted_phrases:
        text = re.sub(re.escape(phrase), '', text, flags=re.IGNORECASE)

    return text
````

````python
import requests
from bs4 import BeautifulSoup
def get_content(url): #'http://book.itep.ru/2/26/ziv_261.htm'

  # –°–æ–≤–µ—Ä—à–∞–µ–º HTTP-–∑–∞–ø—Ä–æ—Å –∫ —Å—Ç—Ä–∞–Ω–∏—Ü–µ
  response = requests.get(url)
  print(url)

  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —É—Å–ø–µ—à–µ–Ω –ª–∏ –∑–∞–ø—Ä–æ—Å (–∫–æ–¥ —Å–æ—Å—Ç–æ—è–Ω–∏—è 200)
  if response.status_code == 200:
      # –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç BeautifulSoup –¥–ª—è —Ä–∞–∑–±–æ—Ä–∞ HTML
      soup = BeautifulSoup(response.text, 'html.parser')

      # –ü–æ–ª—É—á–∞–µ–º –æ—Å–Ω–æ–≤–Ω–æ–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Å—Ç—Ä–∞–Ω–∏—Ü—ã
      # –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –º—ã –∏–∑–≤–ª–µ–∫–∞–µ–º –≤–µ—Å—å —Ç–µ–∫—Å—Ç –∏–∑ —Ç–µ–≥–∞ <body>,
      # –Ω–æ –≤—ã –º–æ–∂–µ—Ç–µ –∞–¥–∞–ø—Ç–∏—Ä–æ–≤–∞—Ç—å —Å–µ–ª–µ–∫—Ç–æ—Ä—ã –ø–æ–¥ –Ω—É–∂–Ω—ã–µ –≤–∞–º —ç–ª–µ–º–µ–Ω—Ç—ã.
      main_content = clean_text(soup.body.get_text())
      return main_content
  else:
      print(f'–û—à–∏–±–∫–∞ –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ç—Ä–∞–Ω–∏—Ü—ã: —Å—Ç–∞—Ç—É—Å –∫–æ–¥ {response.status_code}')
      raise Exception("–¢—ã —á–µ –Ω–∞–¥–µ–ª–∞–ª?")
````

````python
from llama_index.core import Document
from llama_index.core import Settings
from llama_index.core import VectorStoreIndex
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
import chromadb
````

````python
#vars
model_embedding = "intfloat/multilingual-e5-large"
file_with_docs = 'contents.json' #—Å—Å—ã–ª–∫–∞ –Ω–∞ —Ñ–∞–π–ª—ã –¥–∂–µ–π—Å–æ–Ω
name_collection = 'sites' # –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–ª–µ–∫—Ü–∏–∏, –∫—É–¥–∞ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å chroma
````

````python
#–ø–∞—Ä—Å–∏–Ω–≥ json
def create_document_from_json(url,detail,user=1):
  content = details["content"]
  date = details["date"]
  return Document(text=content, metadata={"url": url, "date": date , "user": user})
````

````python
import json
# –ø–æ—Å—Ç—Ä–æ–Ω–∏–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞

db = chromadb.PersistentClient(path="./chroma_db")
chroma_collection = db.get_or_create_collection(name_collection)

# define embedding function
embed_model = HuggingFaceEmbedding(model_name=model_embedding)
Settings.embed_model = embed_model

with open(file_with_docs, 'r', encoding='utf-8') as file:
    data = json.load(file)

documents = []
for url, details in data.items():
    documents.append(create_document_from_json(url ,details))

vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents, storage_context=storage_context, embed_model=embed_model
)
````

````python
from openai import OpenAI
import os
import pandas as pd


def chatgpt_get(user_prompt, system_prompt):
    openai_api_key = '—Ç–æ–∫–µ–Ω'
    if not openai_api_key:
        print("API –∫–ª—é—á OpenAI –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –æ–∫—Ä—É–∂–µ–Ω–∏—è.")
        raise ValueError("API –∫–ª—é—á OpenAI –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –æ–∫—Ä—É–∂–µ–Ω–∏—è.")
    client = OpenAI(api_key=openai_api_key)

    try:
        completion = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {
                    "role": "system",
                    "content": system_prompt},
                {
                    "role": "user",
                    "content": user_prompt
                }
            ]
        )
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞—â–µ–Ω–∏–∏ –∫ OpenAI API: {e}")
        return pd.DataFrame()

    # –ü–æ–ª—É—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    extracted_json = completion.choices[0].message.content.strip()

    return extracted_json
````

````python
system_prompt_question = """
You are an assistant that generates questions based on text batches provided by the user.

Input format:
`{batch_id: "Text of the batch", ...}`

Output format:
`{"Question": "Formulated question based solely on the selected batches", "List of relevant sources": [batch_id, ...]}`

Guidelines:
1. Select up to **one to five relevant batches** that provide the most context for the question.
2. Ensure that the question is clear and can be **only** answered using the selected sources.
3. The question should **not** be answerable using any batches other than those selected.
4. Formulate the question in the language the user provided the text.
"""
````

````python
def update_url_dictionary_with_question_and_texts(url_dict, url, message):
    # –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ —Å–æ–æ–±—â–µ–Ω–∏—è –Ω–∞ —á–∞—Å—Ç–∏
    lines = message.split("\n")
    question = None
    texts = []

    # –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –≤–æ–ø—Ä–æ—Å–∞ –∏ —Ç–µ–∫—Å—Ç–æ–≤
    for line in lines:
        if line.startswith("Question:"):
            question = line.replace("Question:", "").strip()
        elif line.startswith("Texts:"):
            continue  # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º —Å—Ç—Ä–æ–∫—É —Å –∑–∞–≥–æ–ª–æ–≤–∫–æ–º "Texts:"
        elif line.strip().isdigit():
            continue  # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º —Å—Ç—Ä–æ–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –æ–±–æ–∑–Ω–∞—á–∞—é—Ç –Ω–æ–º–µ—Ä–∞ (e.g., "1.", "2.")
        else:
            texts.append(line.strip())

    # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ URL
    if url in url_dict:
        url_dict[url]["question"] = question
        url_dict[url]["relevant_batches"] = texts
    else:
        # –ï—Å–ª–∏ URL –Ω–µ—Ç –≤ —Å–ª–æ–≤–∞—Ä–µ, —Å–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π –∫–ª—é—á —Å –¥–∞–Ω–Ω—ã–º–∏
        url_dict[url] = {"question": question, "relevant_batches": texts}

    return url_dict
````
